```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import spherical_jn, spherical_yn
```

```python
l = 2  
rho = np.linspace(0, 10, 100)  

j = spherical_jn(l, rho)
n = spherical_yn(l, rho)

plt.plot(rho, j, label='Spherical Bessel\'s Function')
plt.plot(rho, n, label='Spherical Neumann Function')
plt.xlabel('rho')
plt.ylabel('Function Value')
plt.legend()
plt.grid(True)
plt.show()
```
