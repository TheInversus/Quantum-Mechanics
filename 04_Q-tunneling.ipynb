```python
def rectangular_double_barrier(w1, w2, a):
    x = np.array([-w1/2, -w1/2, w1/2, w1/2, -w1/2, -w1/2, w1/2, w1/2])
    y = np.array([0, w2, w2, w2+a, w2+a, w2, w2, 0])

    plt.plot([-w1/2, -w1/2], [0, w2], 'k--', linewidth=1) 
    plt.plot([w1/2, w1/2], [0, w2], 'k--', linewidth=1)    

    plt.plot([-w1/2, w1/2], [w2, w2], 'k--', linewidth=1)  

    plt.xlim(-w1/2 - 1, w1/2 + 1)
    plt.ylim(0, w2 + a + 1)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Rectangular Double Barrier')

    plt.show()


w1 = 5
w2 = 3
a = 2
rectangular_double_barrier(w1, w2, a)
```
